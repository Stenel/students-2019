```python
import numpy as np
import matplotlib.pyplot as plt
from numba import jit
```

```python
@jit(nopython = True)
def Solver(n,m,T):
    x = np.linspace(0,1,n)
    t = np.linspace(0,T,m)
    u = np.zeros((n,m))
    u[:,1] = np.sin(np.pi*x)
    u[:,0] = np.sin(np.pi*x)

    C = (t[1]-t[0])/(x[1]-x[0])

    for j in range(2,m):
        for i in range(1,n-1):
            u[i,j] = -u[i,j-2]+2*u[i,j-1]+(u[i+1,j-1]-2*u[i,j-1]+u[i-1,j-1])*(C**2)
    return x, t, u
```

```python
n = 101
m = 1001
T = 4

x,t,u = Solver(n,m,T)
T,X = np.meshgrid(t,x)
u_ = np.cos(np.pi*T)*np.sin(np.pi*X)
```

```python
plt.figure()
plt.contourf(t,x,u,100)
plt.colorbar()
plt.figure()
plt.contourf(t,x,u_,100)
plt.colorbar()
plt.figure()
plt.contourf(t,x,abs(u_-u),100)
plt.colorbar()
print('C = ' + str(C))
```
